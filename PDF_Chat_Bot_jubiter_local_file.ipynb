```python
from google.colab import files
import PyPDF2

# Function to upload the PDF
def upload_pdf():
    uploaded = files.upload()
    for filename in uploaded.keys():
        print(f'Uploaded file: {filename}')
        return filename

# Function to extract text from the uploaded PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text


```

```python
# Upload and extract text from PDF
pdf_path = upload_pdf()
document_text = extract_text_from_pdf(pdf_path)
print(document_text[:1000])  # Print the first 1000 characters of the extracted text
```

```python
from transformers import pipeline

# Load pre-trained Question-Answering model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Function to ask questions from the document text
def ask_question(question, context):
    result = qa_pipeline(question=question, context=context)
    return result['answer']
```

```python
# Example question
question = "Who Is Prime Minister Of India"
answer = ask_question(question, document_text)
print(f"Question: {question}")
print(f"Answer: {answer}")
```
